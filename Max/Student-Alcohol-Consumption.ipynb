```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import plotly.io as pio
```

```python
df_alc = pd.read_csv("student-por.csv")
```

```python
df_alc = df_alc.drop(columns=['G1','G2','school','Medu','Fedu','Mjob','Fjob','freetime','nursery','paid','schoolsup','traveltime','studytime','higher','address'])
```

```python
#Values for students academic reasons to be at school
df_home = df_alc[df_alc['reason']=='home']
df_other = df_alc[df_alc['reason']=='other']
df_course = df_alc[df_alc['reason']=='course']
df_rep = df_alc[df_alc['reason']=='reputation']

#Daily Alcohol consumption for students reasoning for going to school
homeDCount = df_home['Dalc'].value_counts()
otherDCount = df_other['Dalc'].value_counts()
courseDCount = df_course['Dalc'].value_counts()

#Weekend Alcohol consumption for students reasoning for going to school
homeWCount = df_home['Walc'].value_counts()
otherWCount = df_other['Walc'].value_counts()
courseWCount = df_course['Walc'].value_counts()

#Romantic Motivation
df_taken = df_alc[df_alc['romantic']=='yes']
df_single = df_alc[df_alc['romantic']=='no']
takenDCount = df_taken['Dalc'].value_counts()
singleDCount = df_single['Dalc'].value_counts()
```

```python
#Indicator for Motivation and performance
fig = go.Figure()

fig.add_trace(go.Histogram(x=df_home['failures'],histnorm='percent',name='Home Proximity'))
fig.add_trace(go.Histogram(x=df_course['failures'],histnorm='percent',name='Course Choice'))
fig.add_trace(go.Histogram(x=df_other['failures'],histnorm='percent',name='Other Reasoning'))
fig.add_trace(go.Histogram(x=df_rep['failures'],histnorm='percent',name='Reputation Reasoning'))
fig.update_layout(title="Amount of Failed Classes", xaxis=dict(title='# of student failures',
        titlefont_size=16,
        tickfont_size=14),yaxis=dict(title='Percentage of Students',
        titlefont_size=16,
        tickfont_size=14))

pio.write_html(fig,"student-fails.html", full_html=False)
```

```python
fig = go.Figure()

fig.add_trace(go.Histogram(x=df_home['Dalc'],histnorm='percent',name='Home Proximity'))
fig.add_trace(go.Histogram(x=df_course['Dalc'],histnorm='percent',name='Course Choice'))
fig.add_trace(go.Histogram(x=df_other['Dalc'],histnorm='percent',name='Other Reasoning'))
fig.add_trace(go.Histogram(x=df_rep['Dalc'],histnorm='percent',name='Reputation Reasoning'))
fig.update_layout(title="Reasoning for Choosing the University and Alcohol Consumption", xaxis=dict(title='Weekly Drinking Habits, 1-5 (low to high)',
        titlefont_size=16,
        tickfont_size=14),yaxis=dict(title='Percentage of Students',
        titlefont_size=16,
        tickfont_size=14))
pio.write_html(fig,"student-proximty.html", full_html=False)
```

```python
fig = go.Figure()

fig.add_trace(go.Histogram(x=df_single['Dalc'],histnorm='percent',name='Single'))
fig.add_trace(go.Histogram(x=df_taken['Dalc'],histnorm='percent',name='In a Relationship'))
fig.update_layout(title="Students Romantic Status", xaxis=dict(title='Weekly Drinking Habits, 1-5 (low to high)',
        titlefont_size=16,
        tickfont_size=14),yaxis=dict(title='Percentage of Students',
        titlefont_size=16,
        tickfont_size=14))

pio.write_html(fig,"student-romantic.html", full_html=False)
```
